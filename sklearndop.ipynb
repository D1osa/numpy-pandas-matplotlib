Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.
Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
boston.keys()
```

```python
data = boston['data']
data.shape
```

```python
feature_names = boston['feature_names']
feature_names
```

```python
target = boston['target']
target[:10]
```

```python
X = pd.DataFrame(data, columns = feature_names)
X.head()
```

```python
X.info()
```

```python
y = pd.DataFrame(target, columns=['price'])
y.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, shuffle = True, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred.shape
```

```python
check_test = pd.DataFrame({'y_test': y_test['price'],
                          'y_pred': y_pred.flatten(),})
check_test.head(15)
```

```python
check_test['error'] = check_test['y_pred'] - check_test['y_test']
check_test.head()
```

```python
mse1 = (check_test['error']**2).mean()
mse1
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(check_test['y_pred'], check_test['y_test'])
```

```python
from sklearn.metrics import r2_score
```

```python
R2 = r2_score(check_test['y_test'], check_test['y_pred'])
print(R2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/Users/lu/Downloads/creditcard.csv')
df.iloc[:10, [0, 1, 2, 28, 29, 30]]
```

```python
df.value_counts(normalize = True)
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
```

```python
X = df.drop(columns='Class')
X
```

```python
y = df['Class']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

```python
print(y_train.shape)
```

```python
print(y_test.shape)
```

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
grid = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc', cv=3)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred_proba = grid.predict_proba(X_test)
print(y_pred_proba[:10])
```

```python
from sklearn.metrics import roc_auc_score
```

```python
y_pred_proba = y_pred_proba[:,1]
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba, pos_label=1)
```
